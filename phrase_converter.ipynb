```python
import re
```

```python
inputFile = open(r"models/cate/segmentation.txt","r")
tagRemovedFile = open(r"models/cate/segmentationTagRemoved.txt","w")
segmentedPhrasesFile = open(r"models/cate/segmentedPhrases.txt", "w")
```

```python
phraseList = []

def convertPhraseToUnderScore(obj):
    string = obj.group(0)
    string = obj.group(0).replace(" ", "_")[8:-9]
    if string not in phraseList:
        phraseList.append(string)
    return string
result = ""
for item in inputFile:
    result = re.sub(r"<phrase>.*?</phrase>", convertPhraseToUnderScore, item)
    tagRemovedFile.write(result)

for item in phraseList:
    segmentedPhrasesFile.write(item + "\n")
```
